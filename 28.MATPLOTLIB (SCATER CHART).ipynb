```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('employee.csv')
data.head()
```

```python
data.shape
```

```python
plt.rcParams['figure.figsize'] = (9,5)

plt.scatter(data['Age'],
            data['MonthlyRate'])
plt.show()
```
